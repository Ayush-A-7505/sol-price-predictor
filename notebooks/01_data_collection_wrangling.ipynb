```python
import yfinance as yf
import pandas as pd
import numpy as np
import os
```

```python
ticker = "SOL-USD"

raw = yf.download(ticker, start="2020-09-01", end="2024-01-01")

# ✅ Flatten multi-level columns right after download
raw.columns = raw.columns.get_level_values(0)

print("Columns:", raw.columns.tolist())
print(raw.head())
```

```python
os.makedirs("../data", exist_ok=True)
raw.to_csv("../data/raw_sol_data.csv")
print("Clean raw data saved ✅")
```

Data Wrangling steps now
we check for missing values, then feature engineering then the wrangled data is saved

```python
df = raw.copy()

df['Daily_Return']  = df['Close'].pct_change() * 100
df['MA_7']          = df['Close'].rolling(window=7).mean()
df['MA_21']         = df['Close'].rolling(window=21).mean()
df['Price_Range']   = df['High'] - df['Low']
df['Volume_Change'] = df['Volume'].pct_change() * 100
df['Target']        = (df['Close'].shift(-1) > df['Close']).astype(int)

df.dropna(inplace=True)

print(df.shape)
print(df.dtypes)
df.head()
```

```python
df.to_csv("../data/wrangled_sol_data.csv")
print("Wrangled SOL data saved ✅")
print(df.columns.tolist())
```
